```python
import pandas as pd
import json
```

```python
# read data from jsons
data1 = pd.read_json("kl_perfumes_2017_2021/perfume_data.json") 
data2 = pd.read_json("kl_perfumes_2017_2021/perfume_data_2017_plus.json") 
data3 = pd.read_json("kl_perfumes_2017_2021/perfume_data_2020_2021_45_hits_REM.json")
data4 = pd.read_json("perfume_data_jv.json")
data5 = pd.read_json("perfume_data_1920_1923.json")
data6 = pd.read_json("perfume_data_1920_1924.json")
data7 = pd.read_json("perfume_data_cv_2003_2005.json")
data8 = pd.read_json("perfume_data_dv_2008-2012.json")
# data9 = pd.read_json("ozzie_perfume_data_jv.json")
data10 = pd.read_json("perfume_data_chloe_1930_1936.json")
data11 = pd.read_json("perfume_data_JV_1940s.json")
data12 = pd.read_json("perfume_data_KL_1960-67.json")
data13 = pd.read_json("perfume_data_jd.json")
```

```python
# combine all jsons into one df
data_combined = pd.concat([data1, data2, data3, data4, data5, data6, data7, data8, data10, data11, data12, data13], ignore_index=True)
```

```python
data_combined2 = data_combined.copy()
data_combined2
```

```python
# remove duplicates by perfume name 
data_combined2.drop_duplicates(subset=['name'], inplace=True)
```

```python
data_combined2_dict = data_combined2.to_dict()
```

```python
with open ("perfume_data_combined.json", "w") as f:
    json.dump(data_combined2_dict, f)
```
